```python
# 분석에 필요한 패키지 불러오기
import os
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score, roc_auc_score, roc_curve
import statsmodels.api as sm
import matplotlib.pyplot as plt
import itertools
import time
```

```python
data = pd.read_csv('training.csv')
data.head()
```

```python
drop_list = ['홍수사상번호', '연', '월', '일', '시간']
#drop_list = ['홍수사상번호', '연', '월', '일', '시간', '유입량']

Column_list1 = ['1유역평균강수','1강우A','1강우B','1강우C','1강우D','1수위E','1수위D']
Column_list = ['유역평균강수','강우A','강우B','강우C','강우D','수위E','수위D']
#dataX = data.iloc[0:2891, 0:13].drop(drop_list, axis=1)

new_data = data.iloc[0:2891, 0:13].dropna().drop(drop_list, axis=1, inplace=False)
new_data
```

```python
new_data.rename(columns=dict(zip(Column_list1,Column_list)),inplace=True)

#Column_list = ['유역평균강수','강우A','강우B','강우C','강우D','수위E','수위D']
Column_list_english = ['localMeanRain','rainA','rainB','rainC','rainD','waterLevelE','waterLevelD']

new_data.rename(columns=dict(zip(Column_list1,Column_list)),inplace=True)
new_data.rename(columns=dict(zip(Column_list,Column_list_english)),inplace=True)
new_data.rename(columns={'유입량':'inflow'},inplace=True)
new_data
```

```python
feature_columns = list(new_data.columns.difference(['']))
X = new_data[feature_columns]
y = new_data['inflow']
y
```

```python
train_x, test_x, train_y, test_y = train_test_split(X, y, train_size=0.7,test_size=0.3,random_state=42)
print(train_x.shape, test_x.shape, train_y.shape, test_y.shape)
```

```python
reg = LinearRegression()
result = reg.fit(train_x, train_y)
```

```python
pred_Y = reg.predict(test_x)
```
